```python
#Import some libraries that we are going to use later on
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv('/Users/ricardo/Downloads/movies.csv')
```

```python
#A look into the first 5 rows of data
df.head()
```

```python
print("Percentage of missing data")
for col in df.columns:
    pct_missing = np.round(np.mean(df[col].isnull()),5)
    print('{} - {}%'.format(col, pct_missing))
    
print("\nNumbers of rows with missing values")
print(df.isnull().sum())
```

```python
df = df.dropna()

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
pd.set_option('display.max_rows', None)
```

```python
df['year_correct'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df
```

```python
df = df.sort_values(by=["gross"], inplace=False, ascending= False)
```

```python
 #Drop any duplicates
    
df.drop_duplicates()
```

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#Plot Budget vs Gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"},line_kws= {'color':'blue'})
plt.show()
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
    
sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
    
sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs= corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
